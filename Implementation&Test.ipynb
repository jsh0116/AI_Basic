```python
# -*- coding: utf-8 -*-
# -*- coding: ascii -*-
# Scikit-learn 을 활용하여 2개의 데이터셋으로부터 결정 트리 생성
from sklearn.datasets import load_iris 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
#from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import numpy as np
import pandas as pd
from IPython.display import Image
import os
import pydotplus
import pydot
# 시각화 패키지를 이용해 트리를 시각화하여 출력하기 
import matplotlib
import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
# Check open dataset
iris = load_iris()
cancer = load_breast_cancer()
print dir(iris)
print dir(cancer)
print('')
print(iris.DESCR)
print(cancer.DESCR)
```

```python
# Training set과 Testing set 생성
X_train, X_test, y_train, y_test = \
    train_test_split(iris.data, iris.target, test_size=0.25, random_state=42)

Z_train, Z_test, w_train, w_test = \
    train_test_split(cancer.data, cancer.target, test_size=0.25, random_state=42)
```

```python
print(X_train.shape)
print(Z_train.shape)
#Decision Tree Declare
iris_tree = DecisionTreeClassifier(random_state=0)
iris_tree.fit(X_train,y_train)
cancer_tree = DecisionTreeClassifier(random_state=0)
cancer_tree.fit(Z_train,w_train)
accu_train11 = np.sum(iris_tree.predict(X_train) == y_train) / float(y_train.size)
accu_test11 = np.sum(iris_tree.predict(X_test) == y_test) / float(y_test.size)
print ("iris Classification accu on train",accu_train11)
print ("iris Classification accu on test",accu_test11)
accu_train21 = np.sum(cancer_tree.predict(Z_train) == w_train) / float(w_train.size)
accu_test21 = np.sum(cancer_tree.predict(Z_test) == w_test) / float(w_test.size)
print ("Cancer Classification accu on train",accu_train21)
print ("Cancer Classification accu on test",accu_test21)

iris_dot_data1 = export_graphviz(iris_tree, out_file=None, feature_names=iris.feature_names,
                          class_names=iris.target_names, impurity=False, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(iris_dot_data1)
Image(graph.create_png())
```

```python
# A tree for each dataset was created using DecisionTreeClassifier. 
# Try to print the decision tree first without setting the base and the depth of the tree.
cancer_dot_data1 = export_graphviz(cancer_tree, out_file=None, feature_names=cancer.feature_names,
                          class_names=cancer.target_names, impurity=False, filled=True, rounded=True, special_characters=True)
graph_ = pydotplus.graph_from_dot_data(cancer_dot_data1)
Image(graph_.create_png())
```

```python
# change criterion,Tree Depth limit and control random state
iris_treelimit = DecisionTreeClassifier(criterion = 'entropy', max_depth = 3, random_state=42)
cancer_treelimit = DecisionTreeClassifier(criterion = 'gini', max_depth = 4, random_state=42)
# Trainning
iris_treelimit.fit(X_train,y_train)
cancer_treelimit.fit(Z_train,w_train)
accu_train12 = np.sum(iris_treelimit.predict(X_train) == y_train) / float(y_train.size)
accu_test12 = np.sum(iris_treelimit.predict(X_test) == y_test) / float(y_test.size)
print ("iris Classification accu on train",accu_train12)
print ("iris Classification accu on test",accu_test12)
accu_train22 = np.sum(cancer_treelimit.predict(Z_train) == w_train) / float(w_train.size)
accu_test22 = np.sum(cancer_treelimit.predict(Z_test) == w_test) / float(w_test.size)
print ("Cancer Classification accu on train",accu_train22)
print ("Cancer Classification accu on test",accu_test22)
```

```python
iris_dot_data = export_graphviz(iris_treelimit, out_file=None, feature_names=iris.feature_names,
                          class_names=iris.target_names, impurity=False, filled=True, rounded=True, special_characters=True)
graph1 = pydotplus.graph_from_dot_data(iris_dot_data)
Image(graph1.create_png())
```

```python
cancer_dot_data = export_graphviz(cancer_treelimit, out_file=None, feature_names=cancer.feature_names,
                          class_names=cancer.target_names, impurity=False, filled=True, rounded=True, special_characters=True)
graph2 = pydotplus.graph_from_dot_data(cancer_dot_data)
Image(graph2.create_png())
```
